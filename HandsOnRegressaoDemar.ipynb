Sabemos que sklearn apenas recebe arrays como entrada, então precisamos tomar cuidado com a forma que introduzimos os dados

```python
#Vamos fazer uma separação do dataframe para poder realizar a separação em treino e test
label = df['actual']
sequencia_das_colunas = df.columns
df = df.drop('actual', axis = 1)
df.head()
```

```python
#Vamos importar a nossa função de separação em treino e teste
from sklearn.model_selection import train_test_split
```

```python
#Agora vamos montar nossos dados para separar
xtrain, xtest, ytrain, ytest = train_test_split(df.iloc[:,:], label, test_size = 0.25, random_state = 42)
```

Como foi falado nas aulas teóricas, precisamos criar uma base de comparação para os modelos que queremos avaliar.
Várias medidas de erro apresentadas usam como baseline, isto é, uma linha de base, o modelo médio. Para o nosso problema, o baseline utilizado é a média máxima histórica.

```python
#criando o baseline
feature_list = list(df.columns)
feature_list
```

```python
#indexando
baseline_preds = xtest['average']
```

```python
#Fiz uma gambiarra aqui e vamos ver se fica igual o dela
#Importando as métricas para análise
from sklearn import metrics

#Analisando as medidas de erro
print('R2:', metrics.r2_score(ytest, baseline_preds))
print('MSE:', metrics.mean_squared_error(ytest, baseline_preds))
print('RMSE:', np.sqrt(metrics.mean_squared_error(ytest, baseline_preds)))
print('RMSLE:', np.sqrt(metrics.mean_squared_log_error(ytest, baseline_preds)))
print('MAE:', metrics.mean_absolute_error(ytest, baseline_preds))
```

Só para avisar, deu igual o dela, que viagem que ela fez shaushuahsuahsuahs

Agora vamos definir uma função para calcular o MAPE

```python
#Criando a função:
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred)/y_true))*100
```

```python
print('MAPE:', mean_absolute_percentage_error(ytest, baseline_preds))
```

Agora vamos usar uma regressão linear para poder analisar de forma melhor esse df

```python
#importando modelo
from sklearn.linear_model import LinearRegression
```

```python
#Agora vamos treinar o modelo com a separação que já fizemos dos dados
regression = LinearRegression().fit(xtrain, ytrain)
```

```python
y_predict = regression.predict(xtest)
```

```python
#Uma coisa legal que eu gostei que ela faz é montar um df com os valores
#reais e os previstos
df_predict = pd.DataFrame({'Real': ytest, 'Previsto': y_predict})
df_predict.head()
```

```python
#Vamos aplicar as metricas novamente e ver o resultado
print('R2:', metrics.r2_score(ytest, y_predict))
print('MSE:', metrics.mean_squared_error(ytest, y_predict))
print('RMSE:', np.sqrt(metrics.mean_squared_error(ytest, y_predict)))
print('RMSLE:', np.sqrt(metrics.mean_squared_log_error(ytest, y_predict)))
print('MAE:', metrics.mean_absolute_error(ytest, y_predict))
print('MAPE:', mean_absolute_percentage_error(ytest, y_predict))
```

Agora vamos trabalhar com outros modelos   

```python
#Vamos importar o modelo
from sklearn.svm import SVR
```

```python
#Agora vamos setar a regressão e atribuir os hiperparâmetros
reg_svr = SVR(kernel='rbf', epsilon=1.0)
reg_svr.fit(xtrain, ytrain)
```

## Random Forest Regressor
Agora vamos trabalhar com o modelo que a gente deve usar caso não saibamos qual usar, é o melhor modelo para se começar a testar quando não se sabe o que usar!!!

```python
#Vamos importar agora o modelo
from sklearn.ensemble import RandomForestRegressor
```

```python
#Agora que já temos nosso modelo, vamos começar a treina-lo
reg_floresta = RandomForestRegressor(n_estimators=10, random_state=42)
reg_floresta.fit(xtrain, ytrain)
```

```python
#Agora vamos testar nosso modelo e avaliar seu resultado
floresta_predict = reg_floresta.predict(xtest)
df_predict_floresta = pd.DataFrame({'Real':ytest,'Floresta':floresta_predict})
df_predict_floresta.head(10)
```

```python
#Vamos aplicar as metricas novamente e ver o resultado
print('R2:', metrics.r2_score(ytest, floresta_predict))
print('MSE:', metrics.mean_squared_error(ytest, floresta_predict))
print('RMSE:', np.sqrt(metrics.mean_squared_error(ytest, floresta_predict)))
print('RMSLE:', np.sqrt(metrics.mean_squared_log_error(ytest, floresta_predict)))
print('MAE:', metrics.mean_absolute_error(ytest, floresta_predict))
print('MAPE:', mean_absolute_percentage_error(ytest, floresta_predict))
```
