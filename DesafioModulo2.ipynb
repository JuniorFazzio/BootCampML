```
#Vamos começar importando o que eu lembro, pelo menos 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```
df = pd.read_csv("https://pycourse.s3.amazonaws.com/banknote_authentication.txt", header=None, names=['variance', 'sknewness', 'curtosis', 'entropy', 'class'])
#Lendo essa poha de df
```

```
#Vamos dar uma analisada nesse dataframe
print(df.head())
print('')
df.info()
```

```
#Agora que demos uma analisada, vamos separar nossos dados em treino e teste que eu sinceramente, acho que deveria transformar essa class em categorico
df['class'].value_counts()
```

```
#Vou cirar um dataframe alterado para poder mexer
df_alterado = df.copy()
print(df_alterado)
df_alterado['class'] = df_alterado['class'].astype('category')
print(df_alterado.info())
df_alterado['class']
```

```
#Agora vou importar do sklearn o train test split
from sklearn.model_selection import train_test_split
```

```
#Agora que eu importei, vou dividir o df alterado em treino e teste
xtrain, xtest, ytrain, ytest = train_test_split(df_alterado.iloc[:, :-1], df_alterado['class'], test_size = 0.3, random_state = 1)
```

```
#Vamos dar uma verificada se eu consegui separar direito
print(xtrain)
print(ytrain)
```

```
#Agora vamos importar as bibliotecas e os modelos de aprendizagem
from sklearn.neighbors import KNeighborsClassifier
clf_KNN = KNeighborsClassifier(n_neighbors = 5)
```
